```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from sklearn import preprocessing
```

```python
data = pd.read_csv('newData.csv')
```

```python
data[["belongs_to_collection", "adult"]] *= 1
```

```python
language = []
for each in data['spoken_languages']:
    language.append(len(each.split("},")))
data['count_language'] = language
```

```python
data['profit'] = data['revenue'] - data['budget']

success = []
for each in data['profit']:
    if each > 0:
        success.append(1)
    else:
        success.append(0)
data['success'] = success
```

```python
genres = []
genre = []
for each in data['genres']:
    genres.append(str(each).split(","))
for each in genres:
    genre.append(each[0])
data['genre'] = genre
```

```python
data = data[['title', 'original_title', 
             'budget', 'revenue', 'profit', 'success', 'belongs_to_collection', 
             'count_language', 'runtime', 'genre', 'genres', 
             'release_date', 'director', 'actors']]




```

```python
directors = data['director'].tolist()
cleanDirectors = []
for directorString in directors:
    director = directorString.strip('[]')
    director = director.replace("'", "")
    cleanDirectors.append(director.split(',')[0])
data['director'] = cleanDirectors

actors = data['actors'].tolist()
cleanActors = []
for actorString in actors:
    actor = actorString.strip('[]')
    actor = actor.replace("'", "")
    cleanActors.append(actor.split(',')[0])
data['actors'] = cleanActors
```

```python
le = preprocessing.LabelEncoder()
le.fit(data.genre)
data['genre_encode'] = le.transform(data.genre)
```

```python
data = data[['title', 'original_title', 
             'budget', 'revenue', 'profit', 'success', 'belongs_to_collection', 
             'count_language', 'runtime', 'genre', 'genre_encode', 'genres', 
             'release_date', 'director', 'actors']]
```

```python
# Unique actors and directors list

actor_unique_list = list(set(data['actors']))
actors = {'actors': actor_unique_list}
unique_actors = pd.DataFrame(actors)

director_unique_list = list(set(data['director']))
directors = {'directors': director_unique_list}
unique_directors = pd.DataFrame(directors)

unique_actors.to_csv('unique_actors2.csv', index=False, encoding='utf-8-sig')
#unique_directors.to_csv('unique_directors.csv', index=False)

writer = pd.ExcelWriter('unique_directors.xlsx', engine='xlsxwriter')
unique_directors.to_excel(writer, 'Sheet1')
data
```
