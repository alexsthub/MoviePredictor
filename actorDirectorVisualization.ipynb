```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
actorData = pd.read_csv('allActorsWithData.csv')
```

```python
list(actorData)
```

```python
actsSortedByMovies = actorData.sort_values(by=['actorMovieCount'], ascending = False).head(7)
plt.bar(actsSortedByMovies['actorName'],  actsSortedByMovies['actorMovieCount'], align='center', alpha=0.5)
plt.xticks(actsSortedByMovies['actorName'], rotation=45)
plt.title('Top 7 Actors, based off of # of Movie Credits:', fontsize=20)
plt.xlabel('Name of Actor', fontsize=16)
plt.ylabel('# of Credits', fontsize=16)
```

```python
actsSortedByWins = actorData.sort_values(by=['actorAwardWins'], ascending = False).head(7)
plt.bar(actsSortedByMovies['actorName'],  actsSortedByMovies['actorAwardWins'], align='center', alpha=0.5)
plt.xticks(actsSortedByMovies['actorName'], rotation=45)
plt.title('Top 7 Actors, based off of # of Award Wins:', fontsize=20)
plt.xlabel('Name of Actor', fontsize=16)
plt.ylabel('# of Wins', fontsize=16)
```

```python
actsSortedByWins = actorData.sort_values(by=['actorAwardNominations'], ascending = False).head(7)
plt.bar(actsSortedByMovies['actorName'],  actsSortedByMovies['actorAwardNominations'], align='center', alpha=0.5)
plt.xticks(actsSortedByMovies['actorName'], rotation=45)
plt.title('Top 7 Actors, based off of # of Award Nominations:', fontsize=20)
plt.xlabel('Name of Actor', fontsize=16)
plt.ylabel('# of Wins', fontsize=16)
```
