```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
data = pd.read_csv('newMainData.csv')
```

```python
list(data)
```

```python
budgetCounts = data['budget'].value_counts().head(20)


ax = sns.barplot(x=budgetCounts.index,y=budgetCounts.values)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_title('Common Movie Budgets')
plt.tight_layout()
plt.xlabel ('Budgets (in Millions)')
plt.show()

```

```python
profitCounts = data['profit'].value_counts().head(20)


ax = sns.barplot(x=profitCounts.index,y=profitCounts.values)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_title('Common Movie Profits')
plt.tight_layout()
plt.xlabel ('Profit (in Millions)')
plt.show()

# profitCounts['profit']
# data['profit'].head().value_counts()
```
