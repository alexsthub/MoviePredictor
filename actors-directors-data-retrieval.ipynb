```python
# Import libraries

import pandas as pd
from pandas import DataFrame
import json
import ast
```

```python
# Read CSV

creditDS = pd.read_csv("cleanData.csv")
```

```python
# Get the director of a film by finding job = "Director"

json_crew = creditDS['crew']

director_unique_list = []
director_list = []
for movie in json_crew:
    director_movie = []
    crew = ast.literal_eval(movie)
    for crew_member in crew:
        if crew_member['job'] == 'Director':
            director_movie.append(crew_member['name'])
            director_unique_list.append(crew_member['name'])
    director_list.append(director_movie)
director_unique_list = list(set(director_unique_list))
```

```python
# Combines the directors to movie

directorDS = creditDS.drop('crew', 1)
directorDS = directorDS.assign(director = director_list)
```

```python
# Get the first 3 actors name

json_cast = creditDS['cast']

actor_unique_list = []
actor_list = []
for movie in json_cast:
    actor_movie = []
    cast = ast.literal_eval(movie)
    for cast_member in cast[:1]:
        actor_movie.append(cast_member['name'])
        actor_unique_list.append(cast_member['name'])
    actor_list.append(actor_movie)
actor_unique_list = list(set(actor_unique_list))
```

```python
# Combine the directors and first 3 actors to movie

act_dirDS = directorDS.drop('cast', 1)
act_dirDS = act_dirDS.assign(actors = actor_list)
act_dirDS
```

```python
actors = {'actors': actor_unique_list}
unique_actors = pd.DataFrame(actors)
```

```python
directors = {'directors': director_unique_list}
unique_directors = pd.DataFrame(directors)
```

```python
# Exports to csv

act_dirDS.to_csv('act_dirDS.csv', index=False)
unique_actors.to_csv('unique_actors.csv', index=False)
unique_directors.to_csv('unique_directors.csv', index=False)
```
